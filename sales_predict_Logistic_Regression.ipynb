```python
dataset=pd.read_csv('DigitalAd_dataset.csv')
print(dataset.shape)
print(dataset.describe())
print(dataset.head())
print(dataset.isnull())
```

```python
X=dataset.iloc[:, :-1].values
print(X)
```

```python
Y=dataset.iloc[:,-1].values
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train) #fit_transform is used to calculate mean and variance
X_test=sc.transform(X_test) #transform all features using respective mean and variance
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

```python
from numpy.core.multiarray import result_type
#predicting whether the customer is going to buy the item or not
age=int(input("Enter new Customer's Age"))
salary=int(input("Enter new Customer's Salary"))
newCust=[[age,salary]]
result=model.predict(sc.transform(newCust))
print(result)
if result==1:
  print("Customer will buy the Product")
else:
  print("Customer will not buy the Product")

```

```python
#predicting the test data
y_pred=model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),Y_test.reshape(len(Y_test),1)),1))
```

```python
#creating confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(Y_test,y_pred)
print("Confusiopn Matrix")
print(cm)
print("Accuracy Score:{0}%".format(accuracy_score(Y_test,y_pred)*100))
```
